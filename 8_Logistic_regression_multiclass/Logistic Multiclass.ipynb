```python
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
# Print first 5 images
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[67])
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

**Exercise**

Use sklearn.datasets iris flower dataset to train your model using logistic regression. You need to figure out accuracy of your model and use that to predict different samples in your test dataset. In iris dataset there are 150 samples containing following features,

1. Sepal Length
2. Sepal Width
3. Petal Length
4. Petal Width
    
Using above 4 features you will clasify a flower in one of the three categories,    

1. Setosa
2. Versicolour
3. Virginica

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.target
```

```python
lm = LogisticRegression(max_iter=1000)
```

```python
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
lm.fit(X_train_iris, y_train_iris)
```

```python
lm.score(X_test_iris, y_test_iris)
```

```python
y_predicted_iris = lm.predict(X_test_iris)
con_mat = confusion_matrix(y_test_iris, y_predicted_iris)
con_mat
```

```python
sn.heatmap(con_mat, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
